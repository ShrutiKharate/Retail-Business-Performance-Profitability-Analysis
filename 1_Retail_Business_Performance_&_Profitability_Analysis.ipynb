```python
import pandas as pd
```

```python
df = pd.read_csv("/content/Retail_with_categories.csv")
print(df.columns)
df.head(10)
```

```python
df['Inventory_Days'] = df['Total_Cost'] / (df['Cost_Per_Item'] * df['Total_Items'])
df
```

```python
# Check correlation
correlation = df[['Inventory_Days', 'Profit']].corr()
print("Correlation between Inventory Days and Profit:")
print(correlation)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatterplot to visualize the relationship
sns.scatterplot(data=df, x='Inventory_Days', y='Profit')
plt.title("Inventory Days vs Profit")
plt.xlabel("Inventory Days")
plt.ylabel("Profit")
plt.show()
```

```python
segment_perf = df.groupby(['Store_Type', 'Season'])[['Total_Cost', 'Profit']].sum().reset_index()
segment_perf
```

```python
pivot = segment_perf.pivot(index='Store_Type', columns='Season', values='Profit')
sns.heatmap(pivot, annot=True, fmt=".0f", cmap='coolwarm')
plt.title('Profit by Store Type and Season')
```

```python
slow_moving = df[df['Total_Items'] < df['Total_Items'].quantile(0.25)]
overstocked = df[df['Inventory_Days'] > df['Inventory_Days'].quantile(0.75)]
```

```python
slow_moving[['Product', 'Category', 'Total_Items', 'Profit']].sort_values(by='Total_Items').head()
overstocked[['Product', 'Category', 'Inventory_Days', 'Profit']].sort_values(by='Inventory_Days', ascending=False).head()
```

```python
profit_by_cat = df.groupby('Category')[['Profit']].sum().sort_values(by='Profit', ascending=False)
profit_by_cat.plot(kind='bar', figsize=(10, 6), title="Profit by Category")
plt.ylabel("Total Profit")
```

```python
basket_analysis = df.groupby('Transaction_ID')['Total_Items'].sum().describe()
print(basket_analysis)
```

```python
df.groupby('Transaction_ID')['Total_Items'].sum().hist(bins=20)
plt.title("Distribution of Basket Sizes")
plt.xlabel("Items per Transaction")
```
